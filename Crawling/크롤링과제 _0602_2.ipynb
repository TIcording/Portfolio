```python
from selenium import webdriver
import time
import requests
import os

# Chrome WebDriver 경로 설정
driver = webdriver.Chrome()
driver.implicitly_wait(3)

url = 'https://www.musinsa.com/categories/item/004002'
driver.get(url)
time.sleep(3)

# 이미지 저장할 폴더 경로
folder_path = './이미지_폴더'

# 폴더 생성
if not os.path.exists(folder_path):
    os.makedirs(folder_path)

# 이미지 가져오기
image_elements = driver.find_elements(By.CSS_SELECTOR, 'img.lazyload')
for i, image_element in enumerate(image_elements):
    image_url = image_element.get_attribute('data-original')
    image_url = 'https:' + image_url  # 스키마 추가
    image_name = f'image_{i+1}.jpg'
    image_path = os.path.join(folder_path, image_name)

    # 이미지 다운로드
    response = requests.get(image_url)
    with open(image_path, 'wb') as f:
        f.write(response.content)

    print(f'Image {i+1} downloaded successfully.')

# WebDriver 종료
driver.quit()
```

```python
import mysql.connector
import os

# MySQL 연결 설정
config = {
    'user': 'root',
    'password': '1234',
    'host': 'localhost',
    'database': 'shop',
    'raise_on_warnings': True
}

# MySQL 연결 생성
conn = mysql.connector.connect(**config)

# 테이블 생성 쿼리
create_table_query = '''
CREATE TABLE IF NOT EXISTS image_data (
    id INT AUTO_INCREMENT PRIMARY KEY,
    category VARCHAR(255),
    file_path VARCHAR(255)
)
'''

# 테이블 생성
cursor = conn.cursor()
cursor.execute(create_table_query)

# 이미지 저장할 폴더 경로
folder_path = './이미지_폴더'

# 폴더 내 이미지 파일 목록 가져오기
image_files = os.listdir(folder_path)

# 이미지 데이터 삽입
for i, image_file in enumerate(image_files):
    category = '메신저백'  # 카테고리
    file_path = os.path.join(folder_path, image_file)  # 파일 경로

    insert_data_query = '''
    INSERT INTO image_data (category, file_path) VALUES (%s, %s)
    '''
    data = (category, file_path)

    cursor.execute(insert_data_query, data)
    conn.commit()

    print(f"Data {i+1} inserted successfully.")

print("데이터가 MySQL에 저장되었습니다.")

# 연결 종료
cursor.close()
conn.close()
```
