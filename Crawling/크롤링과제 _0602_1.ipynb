### 과제1.
* 바나프레소 홈페이지 매장찾기 들어가서 매장명이랑 주소 크롤링후 엑셀저장



```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException

# Chrome WebDriver 경로 설정

# WebDriver 객체 생성
driver = webdriver.Chrome()
driver.implicitly_wait(3)

url = 'https://banapresso.com/store'
driver.get(url)

# 가게 정보를 담을 리스트
store_info_list = []

# 1부터 11까지의 버튼 클릭
for i in range(1, 12):
    # 버튼 클릭
    button_xpath = f'//li/a[contains(text(), "{i}")]'
    button = driver.find_element(By.XPATH, button_xpath)
    button.click()

    # 가게 정보 파싱
    store_elements = driver.find_elements(By.CLASS_NAME, 'store_name_map')
    for store_element in store_elements:
        store_name = store_element.find_element(By.TAG_NAME, 'i').text
        address = store_element.find_element(By.TAG_NAME, 'span').text.strip()

        store_info = {
            'store_name': store_name,
            'address': address
        }
        store_info_list.append(store_info)

    # "다음" 버튼 클릭
    if i == 5 or i == 10:
        try:
            next_button = driver.find_element(By.XPATH, '//a[contains(text(), "다음")]')
            next_button.click()
            WebDriverWait(driver, 10).until(EC.staleness_of(store_elements[0]))
        except NoSuchElementException:
            break

# 결과 출력
for store_info in store_info_list:
    print('Store Name:', store_info['store_name'])
    print('Address:', store_info['address'])
    print('---')

# WebDriver 종료
driver.quit()
```

```python
import os
import pandas as pd

# 폴더 경로
folder_path = './바나프레소'

# 폴더 생성
if not os.path.exists(folder_path):
    os.makedirs(folder_path)

# 결과를 DataFrame으로 변환
df = pd.DataFrame(store_info_list)

# 엑셀 파일 경로
excel_file_path = os.path.join(folder_path, '바나프레소_가게정보.xlsx')

# 엑셀 파일로 저장
df.to_excel(excel_file_path, index=False)
print("가게 정보가 엑셀 파일로 저장되었습니다.")
```
